```python
import pandas as pd
import numpy as np

# 랜덤 초기화
np.random.seed(42)
```

```python
def make_korean_name():
    sir = ['강', '고', '곽', '구', '권', '금', '기', '김', '남', '남궁',
    '노', '도', '류', '문', '민', '박', '배', '백', '변', '서',
    '성', '손', '송', '신', '심', '안', '양', '엄', '여', '오',
    '우', '원', '유', '윤', '이', '임', '장', '전', '정', '제갈',
    '조', '주', '지', '진', '채', '최', '태', '하', '한', '허',
    '홍', '황', '황보']

    name = [
    '민', '서', '준', '도', '윤', '시', '우', '하', '은', '지',
    '현', '우', '진', '연', '아', '예', '수', '호', '태', '영',
    '재', '정', '성', '승', '동', '민', '채', '원', '린', '유',
    '빈', '안', '안', '솔', '나', '다', '라', '미', '주', '소',
    '희', '경', '혜', '선', '미', '진', '효', '훈', '석', '철',
    '건', '찬', '환', '겸', '웅', '혁', '엽', '율', '안', '온']

    arr_sir = np.array(sir)
    arr_name = np.array(name)

    return np.random.choice(arr_sir) + np.random.choice(arr_name) + np.random.choice(arr_name)
```

```python
names = [ make_korean_name() for _ in range(100) ]
print(names)
```

```python
di = {
    '번호': np.arange(1, 101), # 1 ~ 100까지 순차적인 값들의 배열 생성
    '이름': names,
    '학년': 3,
    '반': np.random.randint(1, 6, size=100),    # 1 ~ 5까지 100개의 데이터 생성
    '국어': np.clip(np.round(np.random.normal(50, 10, size=100),2), 0, 100),
    '수학': np.clip(np.round(np.random.normal(60, 10, size=100),2), 0, 100),
    '영어': np.clip(np.round(np.random.normal(70, 10, size=100),2), 0, 100),
}

df = pd.DataFrame(di)
df
```

```python
# 필터링 하기
cond = df['수학'] > 100.0
df[cond]
```

```python
# 국어 60점 이상, 영어 80점이상인 학생들만 가져오기
cond = (df['국어'] >= 60) & (df['영어'] >= 70)
df[cond]
```

```python
# 데이터 추가하기
# 총점이라는 데이터를 생성하여 컬럼을 추가함
df['총점'] = df['국어'] + df['수학'] + df['영어']
df['평균'] = (df['총점'] / 3).round(2)
df.head()
```

```python
# 100명 전체의 등수 구하기
df['전체등수'] = df['총점'].rank(ascending=False).astype(int)
df.head()
```

```python
# 데이터 정렬하기
df.sort_values(by='전체등수')
```

```python
# 번호 이름 학년 반 국어 수학 영어 총점 평균 전체등수 반등수
# 데이터를 구간화 하기 : 수치형 데이터들을 특정 범위로 나누기

# 절대 범위 나누기 : S(90~100),A(80~89),B(70~79),C(60~69),D(50~59),F(~49) 등급 나누기
# 상대 범위 나누기 : 상위, 중상, 중하, 하위

#등급 부여하기 (절대 범위)
df['등급'] = pd.cut(df['평균'], # 구간화할 데이터 평균
                bins=[0,50,60,70,80,90,100], #구간화할 점수 기준
                labels=['F','D','C','B','A','S'], #부여할 데이터
                right=True
                )

df.head()
```

```python
# 상대구간 나누기
df['분위등급'] = pd.qcut(df['총점'], # 구간화할 데이터 평균
                        q=4, # 몇개 구간으로 나눌지 설정
                        labels=['하위','중하','중상','상위'] #부여할 데이터
                        )

df.head()
```
