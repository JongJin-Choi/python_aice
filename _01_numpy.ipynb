## Copy(복사본)과 View(참조)

```python
# copy(복사본)
arr = np.arange(1, 11)
x = arr.copy()  # 카피를 생성한다.
print(arr)
print(x)    

arr[0] = 100
print(arr)
print(x)    # 원본의 값이 변경되어도 카피본은 변경되지 않는다.
```

```python
# View(참조)
arr = np.arange(1, 11)
x = arr.view()  # 참조를 생성한다.
print(arr)
print(x)

arr[0] = 100    # 원본 배열의 값을 변경
print(arr)
print(x)        # 참조도 같이 바뀐다.

# 참조에서도 값을 변경할 수 있다.
x[-1] = 200
print(arr)      # 참조의 값을 바꾸면 원본도 값이 바뀐다!
print(x)       
```

```python
arr = np.array([1,2,3,4,5])
x1 = arr.copy()
x2 = arr.view()

print(x1.base)  # x1은 복사본이기때문에 참조하는 값이 없다.
print(x2.base)  # x2가 참조하는 원본 배열을 보여준다.
```

```python
# 모양 바꾸기(Reshape)

# 1 ~ 12까지의 값을 갖는 1차원 ndarray를 생성하세요, arange()이용하기
arr = np.arange(1, 13)
print(arr.shape)

# arr을 이용하여 4x3형태의 2차원 배열을 만듬
new_arr = arr.reshape(4, 3) 
print(new_arr)
print(arr)

# 아래와같이 데이터 개수가 일치하지 않으면 변경 불가!
# arr2 = arr.reshape(3,5)

# arr을 이용하여 3x2형태의 2차원 배열을 2개갖는 3차원 배열을 만듬
new_arr = arr.reshape(2, 3, 2) 
print(new_arr)
print(arr)
```

```python
# new_arr은 copy인가? view인가?
print(new_arr.base) # new_arr은 view이다!

new_arr[0] = 100
print(arr)
```

```python
# 데이터들의 순회
for n in arr:
    print(n, end=' ')
```

```python
print(new_arr.shape)

# 3차원 배열의 모양으로 출력하기
'''
100 100 
100 100 
100 100 

7 8
9 10
11 12
'''
for x in new_arr:   # x는 2차원 배열
    for y in x:     # y는 1차원 배열
        for n in y: # n은 y의 아이템
            print(n, end=' ')
        print()
    print('\n')

```

```python
# numpy ndarray의 Iterator ===> nditer(): 몇 차원배열이든지 상관없이 모든 아이템들을 순회한다.
for n in np.nditer(new_arr):
    print(n, end=' ')
```

## Numpy ndarray의 연산과 브로드캐스팅

```python
# 배열 아이템들의 연산
arr = np.arange(1, 11)
print(arr)

# arr배열의 모든 아이템들을 10배하기
for i in range(10):
    arr[i] *= 10
print(arr)

# arr배열의 모든 아이템들을 10배하기
new_arr = arr * 10
print(new_arr)
print(new_arr.base) # new_arr이 복사본인지 View인지 체크하기
```

```python
x = new_arr * arr
print(x)
```

```python
# 2차원 배열의 브로드캐스팅
arr = np.array([[1,2,3],
                [4,5,6]])
print(arr + 10) # 모든 아이템들에 10씩 더하기
print(arr*2)

x = np.array([10, 20, 30])
print(arr + x)

# 2차원 배열들끼리의 더하기
y = np.array([[100],
              [200]])
print(arr + y)
```

```python
# 2차원 배열의 브로드캐스팅
arr1 = np.array([[1,2,3],
                [4,5,6]])

arr2 = np.array([[10,20,30],
                [40,50,60]])
print(arr1*arr2)
```

```python
# 행렬의 곱
arr1 = np.array([[1,2,3],
                [4,5,6]])

arr2 = np.array([[10,20],
                [30,40],
                [50,60]])
print(arr1@arr2)
```

```python
# 유용한 통계 함수들
arr = np.arange(1, 13).reshape(3, 4)
print(arr)

print(f'전체합: {arr.sum()}')    # 전체 합
print(f'전체평균: {arr.mean()}')    # 전체 평균
print(f'{arr.sum(axis=0)}')     # 열방향 합
print(f'{arr.sum(axis=1)}')     # 한 행의 합
print(f'{arr.max(axis=1)}')     # 각 행의 최대값
print(f'{arr.mean(axis=1)}')    # 각 행의 평균
print(f'{arr.mean(axis=0)}')    # 각 열의 평균
```

```python
# 랜덤한 값으로 배열 생성하기 - 테스트 생성을 위하여 사용함

arr = (np.random.random(10) * 100).astype(int)
print(arr)
```

```python
# 평균 60점을 기준으로 표준편차 10, 아이템들의 개수 10칸짜리 배열을 생성
arr = np.random.normal(60, 10, size=10).astype(int)
print(arr)
```

```python
# 배열의 정렬
arr.sort()
print(arr)
```

```python
# 배열의 Join
arr1 = np.random.normal(60, 10, size=10).astype(int)
arr2 = np.random.normal(60, 10, size=10).astype(int)
joined = np.concatenate((arr1, arr2))
joined
```

```python
# 2차원 배열의 concatenate
arr1 = np.random.normal(60, 10, size=10).astype(int).reshape(2,5)   # 2 x 5, 2차원 배열 생성
arr2 = np.random.normal(60, 10, size=10).astype(int).reshape(2,5)   # 2 x 5, 2차원 배열 생성
joined = np.concatenate((arr1, arr2), axis=0)
joined
```

```python
# 2차원배열에 axis=1로하면 열방향 데이터들을 조인한다.
joined = np.concatenate((arr1, arr2), axis=1)
joined
```

```python
# 데이터를 쌓기(join)
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.stack((arr1, arr2), axis=0)
print(arr)
```

```python
# 데이터를 쌓기(join), axis=1, 열방향으로 쌓아서 만들기
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.stack((arr1, arr2), axis=1)
print(arr)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.vstack((arr1, arr2))
print(arr)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.dstack((arr1, arr2))
print(arr)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.hstack((arr1, arr2))   # concatenate(axis=0)와 결과가 같음
print(arr)
```

## Numpy배열에서 데이터 찾기

```python
arr = np.array([1,2,3,4,5,4,4])
result = np.where(arr==4)
print(result)
```

```python
arr = np.arange(1, 11)
result = np.where(arr % 2 == 0) # 값이 짝수인 아이템들만 찾기
print(result)
```

## Numpy에서 데이터 필터링

```python
arr = np.array([10, 20, 30, 40, 50])
filters = [True, False, True, False, True]
result = arr[filters]
print(result)
```

```python
# 클래식한 방식의 필터링
arr = np.arange(1, 11)
# 5이상의 값들만 필터링하기
filters = []
for n in arr:
    if n >= 5:
        filters.append(True)
    else:
        filters.append(False)
print(filters)      
# filtering
result = arr[filters]
print(result)
```

```python
arr = np.arange(1, 11)
# 5이상의 값들만 필터링하기
# filters = arr >= 5  # 조건식(condition)을 filters라는 변수에 저장
result = arr[arr >= 5]

print(filters)
print(result)
```

```python
arr = np.array([5,12,19,21,7,3,40,10])
# 10이상 20이하인 값들만 추출하여 출력하기
result = arr[(arr>=10) & (arr<=20)] # and 대신 &을 사용
print(result)
```

```python
# 
arr = np.random.normal(60, 10, size=12).reshape(4,3).astype(int)
print(arr)
```

```python
print(f'전체합: {arr.sum()}')
print(f'전체 평균: {arr.mean().round(2)}')      # .round(2): 소숫점 2자리까지만 만듬

print(f'한 행의 합: {arr.sum(axis=1)}')     # 각 행들의 합계를 구함, axis=1
print(f'한 행의 평균: {arr.mean(axis=1).round(2)}')     # 각 행들의 합계를 구함, axis=1

# 각 행의 평균이 60.0 이상인 행들만 필터링하기
filters = arr.mean(axis=1) > 60.0
result = arr[filters]
print(result)
```
