- Series : 하나의 열(Column) 데이터를 표현(저장)하기위 설계된 클래스이다.
- DataFrame : 테이블 데이터를 표현(저장)하기위해 설계된 클래스이다.

````python
import pandas as pd
````

## Series

````python
li = [10, 20, 30, 40, 50]
s = pd.Series(li)
print(s)
````

````python
li = ['사과', '배', '바나나', '포도', '수박']
fr = pd.Series(li)
print(fr)
````

````python
# 시리즈 데이터의 접근
print(fr[0])
print(fr[1])
````

````python
di = {0:'사과', 1:'배', 2:'포도'}
print(di[2])
print(di[0])
````

````python
li = ['사과', '배', '바나나', '포도', '수박']
fr = pd.Series(li, index=['a', 'b', 'c', 'd', 'e'])
print(fr)
````

````python
print(fr['c'])
````

````python
di = {'a':'사과', 'b':'바나나', 'c':'포도', 'd':'수박'}
fr = pd.Series(di)
print(fr)
````

````python
# 수정하기: 이미 존재하는 키를 이용하여 데이터를 대입
fr['a'] = '귤'
fr
````

````python
# 입력하기: 없는 키에 데이터를 대입하면 추가
fr['e'] = '딸기'
fr
````

````python
# 삭제하기
fr_drop = fr.drop('a')
fr_drop
````

````python
# 결측치 데이터 처리하기
# 결측치 데이터 삭제하기
df_new = df.dropna()
df_new.info()
````

````python
avg = df['Calories'].mean()
avg

df_new = df.fillna(avg)
df_new.info()
````
